```python
df['datetime'].dtype
```

```python
# 3. Using the date column, create new columns for:
# year, month, day of the week and hour of the day.

# We can use 'dt' to parse the specific year, month, etc
# As the datetime column is already in the correct pandas format

df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day_of_week'] = df['datetime'].dt.day_name() # dt's builtin method, I use it for better readability
df['hour'] = df['datetime'].dt.hour
```

```python
df.head(10)
```

```python
# 4. Rename the values in the season column to spring, summer, fall and winter.

"""
KEY:
season - 1 = spring, 2 = summer, 3 = fall, 4 = winter

Approach:
Using 'map' top replace the integers to strings
"""

df['season'] = df['season'].map({1: 'spring', 2: 'summer', 3: 'fall', 4: 'winter'})
df['season'].tail(10)
```

```python
# 5. Calculate the total number of casual and registered bikes
# rented in the years 2011 and 2012.

# First we collect the bikes in 2011 - 2012
df_2011_2012 = df[df['year'].isin([2011, 2012])]

# Checking
df_2011_2012
```

```python
total_casual = df_2011_2012['casual'].sum()
total_registered = df_2011_2012['registered'].sum()

print('total casual, total registered')
print(total_casual, ',', total_registered)
```

```python
# 6. Calculate the mean of the hourly total rentals count by season.
# Which season has the highest mean ?

# Using groupby method
mean_season = df.groupby('season')['count'].mean()

mean_season
```

```python
## Splitting this question in parts

# 7.1: Are more bikes rented by registered users on working or non-working days?

# First we check the 'workingday' column
df['workingday'].head(5)
```

```python
# We checked for working and non-working day registered rentals
df.groupby('workingday')['registered'].mean()

# Conclusion: Yes, registered users use rentals more on working days.
```

```python
# 7.2: Does the answer differ for non-registered users?

# There's a column 'casual' to which we could do the same, in order to check

df.groupby('workingday')['casual'].mean()
```

```python
df.groupby('workingday')[['registered','casual']].mean()

# Conclusion: Yes, there are less casual users and most of the users are registered.
```

```python
# 7.3: Is the answer the same for both years?

# We can simply add another filter in the previous command
df.groupby(['year', 'workingday'])[['registered','casual']].mean()

# Conclusion: Registered users rent more on working days. As compared to casual
#             users, there are more registered rentals overall.
```

```python
# 8. Which months in the year 2011 have the highest and the lowest
# total number of bikes rented ? Repeat for the year 2012.

# Making a new column for better accessibility
df['total'] = df['registered'] + df['casual']
df
```

```python
# Now, we're going to get the monthly-wise sum of rentals
monthly_sum = df.groupby(['year', 'month'])['total'].sum()
monthly_sum
```

```python
monthly_sum.loc[2011].idxmax()
```

```python
# Now, preparing the final result

print("2011 - Rentals")
print("2011 min", monthly_sum.loc[2011].idxmin())
print("2011 max", monthly_sum.loc[2011].idxmax())
print(" ")
print("2012 - Rentals")
print("2012 min", monthly_sum.loc[2012].idxmin())
print("2012 max", monthly_sum.loc[2012].idxmax())
```

```python
# 9. Which type of weather have the highest and lowest mean of the hourly total
#    rentals count ?

mean_weather = df.groupby('weather')['total'].mean()
mean_weather
```

```python
highest_weather = mean_weather.idxmax()
lowest_weather = mean_weather.idxmin()

print("Highest mean: ", highest_weather)
print("Lowest mean:  ", lowest_weather)
```

```python
# 10. Calculate the correlation between the hourly total rentals count
# and all the numerical columns in the dataset.
# Which column has the highest correlation with the total rentals count?

# We're targetting the numerical cols
numeric_cols = df.select_dtypes(include='number')

# We're using the df['total'] column that we made earlier
correlations = numeric_cols.corr()['total'].sort_values(ascending=False)
correlations
```

```python
# We can see that index 0 is 'total' itself
highestcorr_column = correlations.index[1]
highestcorr_value = correlations.iloc[1]

print("Column with the highest correlation: ", highestcorr_column)
print("Value of the highest correlation: ", highestcorr_value)
```

```python
# 11. Create a new categorical column called day_period, which can take four
# possible values: night, morning, afternoon and evening. These values
# correspond to the following binning of the hour column: 0-6: night, 6-12:
# morning, 12-6: afternoon, 6-24:evening.

# Wanted to use a dictionary here and use map method to complete the task
dict_dayperiod = {
    0: 'night', 1: 'night', 2: 'night', 3: 'night', 4: 'night', 5: 'night',
    6: 'morning', 7: 'morning', 8: 'morning', 9: 'morning', 10: 'morning', 11: 'morning',
    12: 'afternoon', 13: 'afternoon', 14: 'afternoon', 15: 'afternoon', 16: 'afternoon', 17: 'afternoon',
    18: 'evening', 19: 'evening', 20: 'evening', 21: 'evening', 22: 'evening', 23: 'evening'
}

df['day_period'] = df['hour'].map(dict_dayperiod)
df[['hour', 'day_period']].head(10)
```

```python
# Or we can use the labels, bins method
labels = ['night', 'morning', 'afternoon', 'evening']
bins = [0,6,12,18,24]

df['day_period'] = pd.cut(df['hour'], bins=bins, labels=labels, right=False)
```

```python
df[['hour', 'day_period']].head(10)
```

```python
df.columns
```

```python
# 12. Generate a pivot table for the mean of the hourly total rentals count,
# with the index set to the day period and the column set to the working day column.
# What can you observe from the table ?

pivot_table = df.pivot_table(
    values = 'total',
    index='day_period',
    columns='workingday',
    aggfunc='mean'
)

print(pivot_table)
```
